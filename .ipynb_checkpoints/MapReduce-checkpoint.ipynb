```python
data = [
    "big data is important",
    "data science uses big data",
    "big data analytics"
]

mapped = []

for line in data:
    words = line.split()
    for word in words:
        mapped.append((word,1))

print("Mapped Output:")
print(mapped)

word_count = {}

for word,count in mapped:
    if word in word_count:
        word_count[word] += count
    else:
        word_count[word] = count

print("\nWord Count Result:")
for word in word_count:
    print(word,":",word_count[word])
```
