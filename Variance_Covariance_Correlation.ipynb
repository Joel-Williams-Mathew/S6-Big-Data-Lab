```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()
X = iris.data

variance_np = np.var(X,axis=0,ddof=1)
covariance_np = np.cov(X, rowvar=False)
correlation_np = np.corrcoef(X, rowvar=False)

print("Variance :",variance_np)
print("CoVariance :",covariance_np)
print("Correlation :",correlation_np)
```
